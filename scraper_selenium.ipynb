## Installing the necessary packages

```python
#Package installation
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd

# Path to the chromedriver.exe file stored in my local computer
s = Service("C:/Users/QUEST/Desktop/Wode_maya_youtube_scraper/Chrome_driver/chromedriver.exe")
```

## Setting up chrome driver and url link 

```python
# Initialize Chrome WebDriver using the specified service
driver = webdriver.Chrome(service = s)

# Navigate to the YouTube page of WODEMAYA
driver.get("https://www.youtube.com/@WODEMAYA/videos")
```

## By-passing the login page 

```python
# Find the sign-in button using its CSS Selector
sign_in = driver.find_element(By.CSS_SELECTOR, "#yDmH0d > c-wiz > div > div > div > div.v2Yske > div.CqMh6b > div.qqtRac > div.KZ9vpc > form:nth-child(3) > div > div > button > span")

# Click on the sign-in button
sign_in.click()
```

## Locating page elements and inputing them into a dictionary 

```python
# Find all elements on the webpage with the class name "style-scope ytd-rich-grid-media"
videos = driver.find_elements(By.CLASS_NAME, "style-scope ytd-rich-grid-media")

# Create an empty list to store information about each video
video_list = []

# Iterate through each video element found on the webpage
for video in videos:
    title = video.find_element(By.XPATH, './/*[@id="video-title"]').text
    views = video.find_element(By.XPATH, './/*[@id="metadata-line"]/span[1]').text
    when = video.find_element(By.XPATH, './/*[@id="metadata-line"]/span[2]').text

    # Store the information about the current video in a dictionary
    vid_items = {
        'Title': title,
        'Views': views,
        'Posted': when
    }
    
     # Append the dictionary to the video_list
    video_list.append(vid_items)
```

## Converting dictionary into a dataframe and displaying the output

```python
# Create a Pandas DataFrame from the list of dictionaries (video_list)
df = pd.DataFrame(video_list)

# Display the DataFrame
df
```

## Saving dataframe into a .CSV file and saving to local computer

```python
# Export the DataFrame to a CSV file
df.to_csv("C:/Users/QUEST/Desktop/Wode_maya_youtube_scraper/vid_files.csv")
```
